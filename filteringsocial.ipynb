Day 2 , Clearing up the file and filtering 

```python
from nltk.tokenize import  word_tokenize
import nltk
nltk.download('punkt')
```

```python

import csv, re

def split_from_back(string, delim):
    ind = string.rfind(delim)
    first_half = string[0:ind]
    second_half = string[ind + 1:ind+2]
    return [first_half, second_half]

# print(split_from_back("lalal, afsfa, 1", ","))

def clean_non_english(txt):
    txt = re.sub(r'\d{4,8}', ' ', txt)
    txt = re.sub(r'(?:(?:https?|ftp):\/\/)?[\w/\-?=%.]+\.[\w/\-?=%.]+\&\#?[0-9]*', ' ', txt)
    txt = re.sub(r'\@[a-zA-Z0-9]*', ' ', txt)
    txt = re.sub(r'\W+', ' ', txt)
    txt = txt.lower()
    txt = txt.replace("[^a-zA-Z]", " ")
    word_tokens = word_tokenize(txt)
    filtered_word = [w for w in word_tokens if all(ord(c) < 128 for c in w)]
    filtered_word = [w + " " for w in filtered_word]
    return "".join(filtered_word)

with open("fouldata2.csv") as foul:
    data = foul.readlines()
    final = []
    for x in range(1, len(data)):
        classif = (split_from_back(data[x], ","))
        if classif[1] != "0" and classif[1] != "1":
            continue
        else:
            final.append(classif)

for x in range(len(final)):
    if 'b"' in final[x][0]:
        final[x][0] = final[x][0].replace('b"', '')
        # print(final[x][0])
    if "b'" in final[x][0]:
        final[x][0] = final[x][0].replace("b'", '')
        # print(final[x][0])
    if "\\r\\n" in final[x][0]:
        final[x][0] = final[x][0].replace("\\r\\n", '')
        # print("here")
    final[x][0] = clean_non_english(final[x][0])

print(len(final))
```

Prepare Dataset

```python
import pandas as pd
text_data = pd.DataFrame()
text_data["text"] = [final[x][0] for x in range(len(final))]
text_data["label"] = [int(final[y][1]) for y in range(len(final))]

text_data = text_data.sample(frac=1)
```

```python
from sklearn import model_selection, preprocessing, linear_model, naive_bayes, metrics, svm, ensemble
import pandas, xgboost, numpy, textblob, string
from keras.preprocessing import text, sequence
from keras import layers, models, optimizers
```

```python

train_x, valid_x, train_y, valid_y = model_selection.train_test_split(text_data['text'], text_data['label'])
```

```python
def train_model(classifier, feature_vector_train, label, feature_vector_valid, is_neural_net=False):
    classifier.fit(feature_vector_train, label)
    
    predictions = classifier.predict(feature_vector_valid)
    
    if is_neural_net:
        predictions = predictions.argmax(axis=-1)
    
    # return metrics.accuracy_score(predictions, valid_y)
    return metrics.classification_report(predictions, valid_y)
```

#Count Vectorizer for feature engineering

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer(analyzer='word', token_pattern=r'\w{1,}')
count_vect.fit(text_data['text'])

# transform the training and validation data using count vectorizer object
xtrain_count =  count_vect.transform(train_x)
xvalid_count =  count_vect.transform(valid_x)
```

```python
classification_report_bayes = train_model(naive_bayes.MultinomialNB(), xtrain_count, train_y, xvalid_count)
classification_report_linear_model = train_model(linear_model.LogisticRegression(), xtrain_count, train_y, xvalid_count)
classification_report_svm = train_model(svm.SVC(), xtrain_count, train_y, xvalid_count)
classification_report_randomForest = train_model(ensemble.RandomForestClassifier(), xtrain_count, train_y, xvalid_count)
classification_report_XGB = train_model(xgboost.XGBClassifier(), xtrain_count.tocsc(), train_y, xvalid_count.tocsc())
```
